```python
import pandas as pd
```

```python
tabla = pd.DataFrame({'prob': [0.4, 0.3, 0.3],
                      'ret': [-0.1, 0.08, 0.23]
                     })
tabla
```

```python
Er = (tabla['prob'] * tabla['ret']).sum()
Er
```
